## 4. Productos más y menos vendidos

```python
ventas_tienda3 = tienda3.groupby('Producto')['Producto'].count()

mas_vendido_t3 = ventas_tienda3.idxmax()
menos_vendido_t3 = ventas_tienda3.idxmin()

print("Tienda 3 - Más vendido:", mas_vendido_t3)
print("Tienda 3 - Menos vendido:", menos_vendido_t3)
```

## 5. Envío promedio por tienda

```python
envio_prom_tienda1 = tienda['Costo de envío'].mean()
envio_prom_tienda2 = tienda2['Costo de envío'].mean()
envio_prom_tienda3 = tienda3['Costo de envío'].mean()
envio_prom_tienda4 = tienda4['Costo de envío'].mean()

print("Envío promedio tienda 1:", envio_prom_tienda1)
print("Envío promedio tienda 2:", envio_prom_tienda2)
print("Envío promedio tienda 3:", envio_prom_tienda3)
print("Envío promedio tienda 4:", envio_prom_tienda4)
```

## **Visualización de graficas**

```python
import pandas as pd

df_envio = pd.DataFrame({ 'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo promedio de envío': [ envio_prom_tienda1, envio_prom_tienda2, envio_prom_tienda3, envio_prom_tienda4]})
df_envio.set_index('Tienda').plot( kind='bar', figsize=(12,8), color='green', xlabel='Tiendas',ylabel='Costo promedio de envío',
    title='Costo de envío promedio por tienda')
```

```python
df_valoracion = pd.DataFrame({ 'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Valoración promedio': [ valoracion_tienda1, valoracion_tienda2, valoracion_tienda3, valoracion_tienda4 ]})
df_valoracion.set_index('Tienda').plot( kind='line', marker='o', figsize=(12,8), color='red', xlabel='Tiendas',
    ylabel='Calificación promedio', title='Valoración promedio de clientes por tienda')

```

```python
df_ingresos = pd.DataFrame({'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
                            'Ingresos totales': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]})
df_ingresos.set_index('Tienda').plot(kind='barh', figsize=(12,8), xlabel='Tiendas', ylabel='Ingresos', title='Ingresos totales por tienda')
```
